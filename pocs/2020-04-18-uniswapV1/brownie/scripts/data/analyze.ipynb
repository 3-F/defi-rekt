```python
import json
with open('./profits.json', 'r') as f:
    profits = json.load(f)
fig = plt.figure(figsize=[10,8])
ax = fig.add_subplot(1, 1, 1)
ax.set_title('profit by imBTC amount rate & input eth amount')
ax.set_xlim([-1, 10])
ax.set_xlabel('rate (amount0 : total_amount)')
ax.set_xticks(range(-1, 20))
ax.set_ylim([-0.08, 0.15])
ax.set_ylabel('profit (ETH)')
for i, x in enumerate(profits):
    x = [t / 1e18 for t in x]
    ax.plot(x, linestyle='dashed', marker='.', label=f'{2 * i + 1} ETH', lw=1)
    ax.legend(loc=2, bbox_to_anchor=(1.05, 1.0), borderaxespad=0.)
# ax.plot(profits))
```

```python
with open('profits_times.json', 'r') as f:
    profits = json.load(f)

with open('profits_times_acc.json', 'r') as f:
    profits_acc = json.load(f)

profits = [x / 1e18 for x in profits]
profits_acc = [x / 1e18 for x in profits_acc]

fig = plt.figure(figsize=[15, 20])
ax = fig.add_subplot(2, 1, 1)
ax.set_ylabel('profit (ETH)')
ax.set_xlabel('Attack times')
ax.set_title('profit by attck times (100 ETH per attack)')
ax.plot(profits, linestyle='dashed', marker='.', label=f'10 ETH', lw=1)
ax = fig.add_subplot(2, 1, 2)
ax.set_ylabel('profit (ETH)')
ax.set_xlabel('Attack times')
ax.set_title('profit by attck times (Accumulative ETH)')
ax.plot(profits_acc, linestyle='dashed', marker='.', label=f'10 ETH', lw=1)
```
